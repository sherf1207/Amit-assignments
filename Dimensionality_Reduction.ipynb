```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

```python
np.random.seed(0)
x=np.random.rand(100,2)
x[:,1] = 0.5 * x[:, 0] + 0.3 * np.random.rand(100)
```

```python
scaler= StandardScaler()
x=scaler.fit_transform(x)
```

```python
coVariance_matrix = np.cov(x.T)
eigenValues, eigenVectors = np.linalg.eig(coVariance_matrix)

descending_eigens = eigenValues.argsort()[::-1]
eigenValues = eigenValues[descending_eigens]
eigenVectors = eigenVectors[:, descending_eigens]

x_pca = x.dot(eigenVectors)
```

```python
U, S, V = np.linalg.svd(x)
x_pca_svd = x.dot(V)
```

```python
# Step 5: Plot results
fig, ax = plt.subplots(1, 3, figsize=(18, 5))
# Original data plot
ax[0].scatter(x[:, 0], x[:, 1], c='b', label='Original Data')
ax[0].set_title('Original Standardized Data')
ax[0].grid()
ax[0].legend()

# PCA with Eigenvalue Decomposition
ax[1].scatter(x_pca[:, 0], x_pca[:, 1], c='g', label='PCA (Eigen Decomposition)')
ax[1].set_title('PCA using Eigenvalue Decomposition')
ax[1].grid()
ax[1].legend()

# PCA with SVD
ax[2].scatter(x_pca_svd[:, 0], x_pca_svd[:, 1], c='r', label='PCA (SVD)')
ax[2].set_title('PCA using SVD')
ax[2].grid()
ax[2].legend()

plt.show()
```
