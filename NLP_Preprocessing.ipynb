```python
# replacing "/" with "per"
import re
df['description'] = df['description'].apply(lambda x: re.sub(r'/', ' per ', x))
df['description']
```

```python
# removing any spectial characters
import re
df['description'] = df['description'].apply(lambda x: re.sub(r'[^a-zA-Z0-9\s]', '', x))
df['description']
```

```python
for i in df['salary_range']:
    df['salary_range']=re.sub(r'-',' to ',i)
df['salary_range'] = df['salary_range'].apply(lambda x: re.sub(r'[^a-zA-Z0-9\s]', '', x))
df['salary_range']
```

```python
import re
text= """
Hey, check this out! Visit https://www.example.com/ for the latest updates on #AI! 🙌🧠 The #machinelearning community is growing fast, but don't forget to always remember, "Data is key!" 😉 And don't miss the *amazing* event happening @ https://www.anotherexample.org/ in 2024!! 😊

#Welcome to the world of NLP, where text is 💬 & emotions are real! Processing text like "this" is fun, but sometimes it's hard to know what to remove. Do you need https://www.randomsite.net/ for your next project? 💻 Or maybe just go to https://www.testpage.io/? It's all good, right?

🚨🎉📊 Don't let those stop words slow you down... they’re everywhere, like "and", "the", "is", "it", "in", "to", "that" 🙈! Just #stayfocused and clean your text! Don't forget to filter out: "he", "she", "we", "you" and so on. 🌟

Moreover, take a look at these examples:
👉 www.somepage.com?query=hello&lang=en
👉 https://www.website1234.com/api?key=xyz

Happy preprocessing! 🧑‍💻✨
"""
```

```python
email_regex = r"https://[^\s]+"
emails = re.findall(email_regex, text)
print("\nExtracted Email Addresses:", emails)
```

```python
 url_clean=re.sub(r"https://[^\s]+","",text)
url_clean=re.sub(r"www.[^\s]+","",url_clean)
print("\nText after Removing url:\n", url_clean)
```

```python
special_characters_clean=re.sub("[^a-zA-Z0-9\s]","",url_clean)
print("\nText after Removing Special Characters:\n", url_clean)
```

```python
# removing emojis
emoji_pattern = re.compile(
    "[\U0001F600-\U0001F64F]"  # Emoticons
    "|[\U0001F300-\U0001F5FF]"  # Symbols & pictographs
    "|[\U0001F680-\U0001F6FF]"  # Transport & map symbols
    "|[\U0001F1E0-\U0001F1FF]"  # Flags
    "|[\U00002702-\U000027B0]"  # Additional emoji ranges
    "|[\U000024C2-\U0001F251]"  # Miscellaneous symbols
    "|[\U0001f926-\U0001f937]"  # Extended emoji ranges
    "|[\U0001F91D-\U0001F92C]"  # Extended emoji ranges
    "|[\U0001F600-\U0001F64F]"
)

emoji_clean = emoji_pattern.sub(r"", special_characters_clean)
print(emoji_clean)
```

```python
all_clean = re.sub(r"  ", " ", emoji_clean)
all_clean = re.sub(r"   ", " ", all_clean)
all_clean = re.sub(r"    ", " ", all_clean)
all_clean = re.sub(r"     ", " ", all_clean)

print("\nUpdated Text after Replacement:\n", all_clean)
```

```python
text= re.sub(r"AI","Artificial intelligence",all_clean)
text= re.sub(r"NLP","Natural Language Processing",text)

print(text)
```

```python
text= re.sub(r"\n","",text)

print(text)
```
