```python
import numpy as np
import pandas as pd
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense, Flatten
from tensorflow.keras import regularizers
from sklearn.preprocessing import StandardScaler
```

```python
df=pd.read_csv('/content/creditcardFraud.csv')
df
```

```python
df.isna().sum()
```

```python
df.duplicated().sum()
```

```python
df=df.drop_duplicates()
```

```python
from sklearn.model_selection import train_test_split
x = df.drop(columns=['class'])
y = df['class']
scaler = StandardScaler()
features_scaled = scaler.fit_transform(x)
x_train, x_test, y_train, y_test = train_test_split(features_scaled, y, test_size=0.2, random_state=42)
```

```python
print(x_train.shape)
```

# applying the autoencoder

```python
input_layer = Input(shape=(30,))

encoded = Dense(15, activation='leaky_relu')(input_layer)
encoded = Dense(7, activation='leaky_relu')(encoded)

decoded = Dense(15, activation='leaky_relu')(encoded)
output_layer = Dense(30, activation='sigmoid')(decoded)

autoencoder = Model(input_layer, output_layer)

autoencoder.compile(optimizer='adam', loss='binary_crossentropy')
```

```python
autoencoder.fit(x_train, x_train,epochs=20, batch_size=256, shuffle=True, validation_data=(x_test, x_test))
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
svc_clf = SVC()
svc_clf.fit(x_train, y_train)
svc_predictions = svc_clf.predict(x_test)
accuracy=accuracy_score(y_test, svc_predictions)*100
```

```python
print("YOUR PREDICTIONS IS {:.1f}".format(accuracy),"ACCURATE")
```
